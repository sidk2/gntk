```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
%matplotlib inline
depths = np.array([1, 2, 4, 8, 16, 32])
```

```python
gntk_cora = np.delete(np.load('GCN_cora.npy')[0], 4)
gntk_jk_cora = np.delete(np.load('GCN_jk_cora.npy')[0], 4)
gntk_skip_cora = np.delete(np.load('GCN_skip_cora.npy')[0], 4)
ssgctk_cora = np.delete(np.load('SSGC_cora.npy')[0], 4)
```

```python
print(gntk_cora)
```

```python
gcn_cora = np.load('gcn_network_cora.npy')[:, 1]
ssgc_cora = np.load('ssgc_network_cora.npy')[:, 1]
```

```python
rc('font', size=12)
plt.figure(figsize=(5, 3))
plt.plot(depths, gntk_cora, label='GNTK', marker='o')
plt.plot(depths, gntk_jk_cora, label='GNTK with jumping knowledge', marker ='s')
plt.plot(depths, gntk_skip_cora, label='GNTK with skip connections', marker = '^')
plt.plot(depths, ssgctk_cora, label='SSGC NTK', marker = 'd')

plt.xticks(depths)
plt.xlim([depths[0], depths[-1]])

plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.title('Accuracy of Neural Tangent Kernels on Cora')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(depths, gntk_cora, label='GNTK', marker='o')
plt.plot(depths, gcn_cora, label='GCN', marker = 's')
#plt.plot(depths, gntk_skip_cora, label='GNTK with skip connections')
#plt.plot(depths, accuracy_net4, label='Network 4', marker='d')

plt.xticks(depths)
plt.xlim([depths[0], depths[-1]])

plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.title('Comparison of GCN to corresponding GNTK')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(5, 3))
plt.plot(depths, ssgctk_cora, label='SSGC NTK', marker='o')
plt.plot(depths, ssgc_cora, label='SSGC Network', marker = 's')
#plt.plot(depths, gntk_skip_cora, label='GNTK with skip connections')
#plt.plot(depths, accuracy_net4, label='Network 4', marker='d')

plt.xticks(depths)
plt.xlim([depths[0], depths[-1]])

plt.xlabel('Depth')
plt.ylabel('Accuracy')
plt.title('Comparison of SSGC network to SSGC NTK')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
